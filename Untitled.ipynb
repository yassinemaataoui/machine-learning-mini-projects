### bibliothèque numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

### déclarer un tableau numpy contenant les données

```python
X = np.array([[6], [8], [10], [14], [18]])
y = [7, 9, 13, 17.5, 18]
```

### Affichage des données

```python
plt.figure()
plt.title("Pizza prices plotted against sizes")
plt.xlabel("Sizes in cms")
plt.ylabel("Prices in euros")
plt.scatter(X, y, color='green')
plt.grid(True)
plt.show()
```

### la bibliothèque sklearn et en particulier le modèle de régression linéaire

```python
from sklearn.linear_model import LinearRegression
```

### Création et entraînement du modèle

```python
model = LinearRegression()
model.fit(X, y)
```

### Prédiction sur les mêmes données pour afficher la droite

```python
y_pred = model.predict(X)
```

### Affichage des données et de la droite de régression

```python
plt.figure()
plt.title("Pizza prices plotted against sizes")
plt.xlabel("Sizes in cms")
plt.ylabel("Prices in euros")
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.grid(True)
plt.show()
```

### Affichage des coefficients

```python
alpha = model.coef_[0]
beta = model.intercept_
print(f"Modèle : y = {alpha:.2f} * x + {beta:.2f}")
```

### Calcul de la somme résiduelle des carrés (RSS)

```python
RSS = ((y - y_pred) ** 2).sum()
print(f"RSS = {RSS:.2f}")
```

### Calcul manuel de alpha et beta

```python
y = np.array(y)
x_mean = X.mean()
y_mean = y.mean()
```

### Variance de X (avec correction ddof=1)

```python
var_x = np.var(X, ddof=1)
```

### Covariance entre X et y

```python
cov_xy = np.cov(X.T, y)[0, 1]
```

### Coefficients

```python
alpha_manual = cov_xy / var_x
beta_manual = y_mean - alpha_manual * x_mean
```

```python
print(f"Alpha (pente) = {alpha_manual:.2f}")
print(f"Beta (ordonnée à l'origine) = {beta_manual:.2f}")
```

### Données de test

```python
X_test = np.array([[8], [9], [11], [16], [12]])
y_test = np.array([11, 8.5, 15, 18, 11])
```

### Prédiction sur le test

```python
y_test_pred = model.predict(X_test)
```

### Calcul du R²

```python
r2_score = model.score(X_test, y_test)
print(f"R² sur le test = {r2_score:.2f}")
```
